```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def request_await(url):
    soup = BeautifulSoup(requests.get(url).text, 'html.parser')
    return soup
```

```python
url = 'https://www.indiatoday.in/top-stories'
common_url = 'https://www.indiatoday.in'
```

```python
soup = request_await(url)
top_stories = soup.find_all('div',{'class':'catagory-listing'})
articles_list = []
for report in top_stories:
    image = report.find('img')['src']
    title = report.find('a').text
    story_url = common_url + report.find('a')['href']
    brief = report.find('p').text
    soup2 = request_await(story_url)
    descriptions = soup2.find_all('div',{'class':'description'})
    
    article = []
    for description in descriptions:
        paragraphs = description.find_all('p')
        for paragraph in paragraphs:
            content = paragraph.text
            article.append(content)
    #print(article,'\n\n')
    articles_list.append([title, story_url, brief, article, image])
```

```python
df = pd.DataFrame(articles_list, columns=['Title', 'Story Url', 'Brief Intro', 'Paragraph', 'Image Url'])
```
