```
class dog(object):
  def __init__(self,name):
    self.name = name
    
  def greet(self , loud = False):
    if loud:
      print("hii {}".format(str(self.name.upper()))
    else:
      print("hello {}".format(str(self.name()))
  
  
g = dog('simely')
g.greet();
g.greet(loud=True)
 
```
