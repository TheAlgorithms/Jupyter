```
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=data[:,0]
y=data[:,1]

plt.figure(figsize=(8,6))
plt.plot(x,y,'r')

plt.xlabel('x')
plt.ylabel('y')

plt.title(r"Plot of y")
plt.show()
```
