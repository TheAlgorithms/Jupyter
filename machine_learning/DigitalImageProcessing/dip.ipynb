```python
# 5 --------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='monospace',size='9')
plt.rc('axes',unicode_minus='False')

mu=0         
sigma=0.8  
x=np.linspace(-3,3,60)  
y=np.exp((-(x-mu)**2)/(2*(sigma**2)))/(np.sqrt(2*np.pi)*sigma)  

plt.plot(x,y,"b-",) 
plt.grid(True)
plt.show()

```
