```python
import numpy as np
import matplotlib.pyplot as plt

# Increase size of plot in jupyter

plt.rcParams["figure.figsize"] = (10,7)
```

```python
# creating the dataset without importing CVS

data={'Mustang':105932,'Camaro':72705,'Challenger':64478,
      'Veloster':30052,'Corvette':29995,'Model S':29421, 'MX':98901,
      '911':9465,'XE':6656,'370Z':5913}

car = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (30, 12))
```

```python
# creating the bar plot

plt.bar(car, values, color ='maroon',
        width = 0.4)
 
plt.xlabel("Car Names")
plt.ylabel("No. Of Customers Brought")
plt.title("Best Selling Sport Cars In US")
plt.show()
```
