```python
from skimage import io
photo = io.imread('my.png')
type(photo)
```

```python
photo.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
plt.imshow(photo[::-1]) # Reverse image
```

```python
plt.imshow(photo[:,::-1]) # Mirror image
```

```python
plt.imshow(photo[100:300, 100:300])
```

```python
plt.imshow(photo[::2, ::2])  #half
```

```python
photo
photo_sin = np.sin(photo)  #sine 
photo_sin
```

```python
print(np.sum(photo))    #sum of all the element in photo
print(np.prod(photo))   #product
print(np.mean(photo))   #mean
print(np.std(photo))    #standard deviation
print(np.var(photo))    #variance
print(np.min(photo))    # min value
print(np.max(photo))    #max value
print(np.argmin(photo)) #index value of min
print(np.argmax(photo)) #index value of max
```
