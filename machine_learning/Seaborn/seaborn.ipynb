# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
a = sns.load_dataset("flights")
sns.relplot(x = "passengers",y = "month",data =a ) # scatter plot
```

```python
a = sns.load_dataset("flights")
sns.relplot(x = "passengers",y = "month",hue="year",data =a )
```

```python
b = sns.load_dataset("tips")
sns.relplot(x ="time",y = "tip",data = b,kind= "line") # line plot
```

```python
sns.catplot(x="day",y="total_bill", data=b)
```

```python
sns.catplot(x="day",y="total_bill", data=b, kind="violin")
```

```python
sns.catplot(x="day",y="total_bill", data=b, kind="box")
```

# Bivariate

```python
a = sns.load_dataset("flights")
sns.jointplot(x = "passengers",y = "year",data =a,kind='kde')
```

```python
b = sns.load_dataset("tips")
sns.jointplot(x ="time",y = "tip",data = b, kind='')
```

# Multiplot Function

```python
a = sns.load_dataset("iris")
b = sns.FacetGrid(a, col='species')
b.map(plt.hist, 'sepal_length')
```

```python
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

# Plot-Aesthetics

```python
sns.set(style='darkgrid')         # background
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='dark')         # background
a = sns.load_dataset("flights")
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='white', color_codes = True)        
a = sns.load_dataset("tips")
sns.boxplot(x='day',y="total_bill",data=a)
```

```python
sns.set(style='white', color_codes = True)        
a = sns.load_dataset("tips")
sns.boxplot(x='day',y="total_bill",data=a)
sns.despine(offset=10,trim=True)
```
