# Special Functions

1. Exponential Functions

```python
from scipy import special

a = special.exp10(2)  # 10 pow(2)
print(a)
```

```python
b = special.exp2(6)
print(b)
```

2. Trignometric Functions

```python
c = special.sindg(90)
print(c)
```

```python
d = special.sindg(45)
print(d)
```

```python
d = special.cosdg(90)
print(d)
```

```python
d = special.cosdg(0)
print(d)
```

```python
d = special.tandg(45)
print(d)
```

# Integration Functions

1. General Intergration

```python
from scipy import integrate
help(integrate.quad)
```

```python
i = scipy.integrate.quad(lambda x: special.exp10(x),0,1)
print(i)
```

2. Double intergration

```python
e = lambda x,y: x*y**2
f = lambda x: 1
g = lambda x:-1
integrate.dblquad(e,0,2,f,g)
```

# Fourier Transformations

```python
from scipy.fftpack import fft,ifft
import numpy as np 
arr = np.array([1,2,3,4,5])
fft = fft(arr)      #fourier transform
ifft = ifft(arr)    #inverse fourier transform
print("array = " ,arr)
print("\n fourier transform = ",fft)
print("\n inverse fourier transform = ",ifft)
```

# Linear Algebra 

```python
from scipy import linalg
a = np.array([[5,6],[8,4]])
b = linalg.inv(a)
print(b)
```

#  Interpolation Functions

```python
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(5,20)
y = np.exp(x/3.0)
f = interpolate.interp1d(x,y)

x1 = np.arange(6,12)
y1 = f(x1)               # use interpolation function returned by 'interp1d'
plt.plot(x,y, 'o' ,x1,y1, '--')
```
