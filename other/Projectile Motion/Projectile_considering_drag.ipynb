```python
#Code for AMC Group Project

#With Drag





import matplotlib.pyplot as plt
import math


#coef of restitution e

e=0.8


#Variable list

#D,Density,C,A (NOTE :A is projection surface area of ball not actual surface area)




#Not very sure about value of drag coeffecient ,  Typical values of C for balls are in the range from 0.2 to 1.0.

C=0.13 #assuming soccerball

massBall=0.45

RHO=1.225

A=0.03976

#D is (RHO*C*A)/2

D=(RHO*C*A)/2


#Rho of air at stp is 1.225 kg/m^3

#A for the given ball (Diameter 225mm) = 
#Density is P(RHO) Of air

def projectile(xi,yi,vel_x,vel_y):
    g=9.81

    # set the time interval as 0.1 s
    delta_t = 0.000001
    v=vel_y
    fx=((-1*D*v*vel_x)/(massBall))
    fy=((-1*D*v*vel_y)/(massBall))

    ax=fx
    ay=g+fy

    x_traj = [xi]
    y_traj = [yi]
    # Calculate position at t = 0.1
    while (yi>0): 
        v=math.sqrt((vel_x*vel_x)+(vel_y*vel_y))
        xi = x_traj[-1] + vel_x*delta_t
        yi = y_traj[-1] + vel_y*delta_t

        # Calculate vel_x and vel_y 
        vel_y = vel_y - ay*delta_t
        vel_x=vel_x + ax*delta_t
        # add new position to trajectory data
        x_traj.append(xi)
        y_traj.append(yi)
    return xi ,yi, vel_y, x_traj, y_traj





output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj=output[3]
y_traj=output[4]

if e>=0.7:
	number_of_runs=12
else:
	number_of_runs=6


for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj=x_traj+output[3]
    y_traj=y_traj+output[4]
    vel_y= -1*vel_y*e

#...............................................................................

e=0.6

output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj1=output[3]
y_traj1=output[4]

if e>=0.7:
	number_of_runs=12
else:
	number_of_runs=6


for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj1=x_traj1+output[3]
    y_traj1=y_traj1+output[4]
    vel_y= -1*vel_y*e



#...............................................................................


e=0.4


output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj2=output[3]
y_traj2=output[4]

if e>=0.7:
	number_of_runs=12
else:
	number_of_runs=6


for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj2=x_traj2+output[3]
    y_traj2=y_traj2+output[4]
    vel_y= -1*vel_y*e



#...............................................................................

e=0.2


output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj3=output[3]
y_traj3=output[4]

if e>=0.7:
	number_of_runs=12
else:
	number_of_runs=6


for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj3=x_traj3+output[3]
    y_traj3=y_traj3+output[4]
    vel_y= -1*vel_y*e



#...............................................................................



e=1


output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj4=output[3]
y_traj4=output[4]

if e>=0.7:
	number_of_runs=12
else:
	number_of_runs=6


for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj4=x_traj4+output[3]
    y_traj4=y_traj4+output[4]
    vel_y= -1*vel_y*e


#...............................................................................



plt.figure(figsize=(20,10))
plt.xlim([0, x_traj[-1]+1])
plt.ylim([0, 5])
plt.plot(x_traj, y_traj, x_traj1, y_traj1,x_traj2,y_traj2,x_traj3,y_traj3)
plt.show()
```
