```python
#Code for AMC Group Project

#Without Drag

import matplotlib.pyplot as plt
import math

#coef of restitution e

e=0.8


def projectile(xi,yi,vel_x,vel_y):
    g=9.81
    # set the time interval as 0.1 s
    delta_t = 0.000001

    x_traj = [xi]
    y_traj = [yi]
    # Calculate position at t = 0.1
    while (yi>0): 
        xi = x_traj[-1] + vel_x*delta_t
        yi = y_traj[-1] + vel_y*delta_t

        # Calculate vel_x and vel_y 
        vel_y = vel_y - g*delta_t
       
        # add new position to trajectory data
        x_traj.append(xi)
        y_traj.append(yi)
    return xi ,yi, vel_y, x_traj, y_traj

output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj=output[3]
y_traj=output[4]


if e>=0.7:
    number_of_runs=12
else:
    number_of_runs=6

for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj=x_traj+output[3]
    y_traj=y_traj+output[4]
    vel_y= -1*vel_y*e



#...............................................................................

e=0.6



output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj1=output[3]
y_traj1=output[4]


if e>=0.7:
    number_of_runs=12
else:
    number_of_runs=6

for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj1=x_traj1+output[3]
    y_traj1=y_traj1+output[4]
    vel_y= -1*vel_y*e


#...............................................................................

e=0.4


output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj2=output[3]
y_traj2=output[4]


if e>=0.7:
    number_of_runs=12
else:
    number_of_runs=6

for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj2=x_traj2+output[3]
    y_traj2=y_traj2+output[4]
    vel_y= -1*vel_y*e


#...............................................................................

e=0.2


output=projectile(0,4,1,0)
xi =  output[0]
vel_y= output[2]

vel_y= -1*vel_y*e
x_traj3=output[3]
y_traj3=output[4]


if e>=0.7:
    number_of_runs=12
else:
    number_of_runs=6

for i in range(number_of_runs):
    output=projectile(xi,0.000000001,1,vel_y)

    xi =  output[0]
    vel_y= output[2]
    x_traj3=x_traj3+output[3]
    y_traj3=y_traj3+output[4]
    vel_y= -1*vel_y*e


#...............................................................................


plt.figure(figsize=(20,10))
plt.xlim([0, x_traj[-1]+1])
plt.ylim([0, 5])
plt.plot(x_traj,y_traj,x_traj1,y_traj1,x_traj2, y_traj2, x_traj3,y_traj3)
plt.show()
```
